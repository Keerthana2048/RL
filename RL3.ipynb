```python
class RewardTracker:
  def __init__(self):
    self.total_reward = 0

  def add_reward(self, reward):
    self.total_reward += reward
    print(f"Added reward {reward}. Total now = {self.total_reward}")


#example usage
tracker = RewardTracker()
tracker.add_reward(5)
tracker.add_reward(-2)
```

```python
import random

class Agent:
  def __init__(self, actions ):
    self.actions = actions

  def choose_action(self):
    action = random.choice(self.actions)
    print(f"Agent chose action : {action}")
    return action

#usage
agent = Agent(["left","right","up","down"])
agent.choose_action()
agent.choose_action()
```

```python
class Environment:
  def __init__(self):
    self.state = 0
  def step(self, action):
    if action == "right" :
      self.state += 1
    elif action == "left" :
      self.state -= 1
    reward = 1 if self.state == 2 else 0
    print(f"State: {self.state}, Reward: {reward}")
    return self.state, reward

#usage
env = Environment()
env.step("right")
env.step("right")
env.step("left")
```
